```python
def remove_outliers(df, column, threshold=3):
    mean= df[column].mean()
    std_dev = df[column].std()
    
    lower_bound = mean - threshold * std_dev #PEMDAS
    upper_bound = mean + threshold * std_dev
    # remove outliers
    
    df = df[(df[column]>= lower_bound) & (df[column]<= upper_bound)]
    return df
```

```python
df_cleaned = remove_outliers(df, 'age')
```

```python
#treating outliers using IQR (interquartilr range)
path = r"C:\Users\DOYIN\Downloads\winequality-red.csv"
wine_data = pd.read_csv(path)

wine_data.head()
```

```python
#codes for calculating number of outliers
def Outliers(df, col):
    IQ1 = df[col].quantile(0.25)
    IQ3 = df[col].quantile(0.75)
    IQR = IQ3 - IQ1
    
    lower_bound = IQ1 - 1.5*IQR
    upper_bound = IQ3 + 1.5*IQR
    index = df.index[(df[col] < lower_bound) | (df[col]>upper_bound)]
    return index
```

```python
index
```

```python
wine_data.index
```

```python
wine_data.shape
```

```python
index = []
for i in wine_data.columns:
    index.extend(Outliers(wine_data, i)) #extend makes the list bigger
index = set(index)  #set helps remove duplicates

print('Total number of outliers are {}'.format(len(index)))

#dropping all the outliers
wine_data.drop(index, inplace=True, axis=0)
wine_data.shape
```
