```python
# importing dependencies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Set display option 
pd.set_option('display.max.columns', None)
```

```python
#data collection and data prepocessing
sonar_data = pd.read_csv('./dataset/sonar.csv',header=None)
```

```python
# checking how many rock and mine is in the 60th column that holds the rock and mine
sonar_data[60].value_counts()
```

M ---> Mine </br>
R ---> Rock

```python
sonar_data.groupby(60).mean()
```

Seperating the data from the labels

```python
#seperating data and label
X = sonar_data.drop(columns=60, axis=1)
Y = sonar_data[60]
```

```python
print(X)
print(Y)
```

Seperating Training and Test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, stratify=Y, random_state=1)
```

X_train == training data </br>
X_test == testing data </br>
Y_train == label of the training data </br>
Y_test == label of the test data</br>
test_size=0.1 means we need 10% of the dat to be test data it can be 0.1 or 0.2</br>
stratify=Y means we want to split the data into equal portion, i.e equal number of Rock and Mine.</br>
random_state=1 is used to split the data in a particular order(i.e to reproduce the code as it is).

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
print(Y.shape,Y_train.shape,Y_test.shape)
```

```python
X_train
```

```python
Y_train
```

Model trainig  ---> Logistic Regression 

```python
model = LogisticRegression()
```

```python
#training the logistic regression model with the training data
model.fit(X_train, Y_train)
```

Model Evaluation

```python
# Accuracy on Training Data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on training data: ', training_data_accuracy)
```

```python
# Accuracy on test Data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy on test data: ", test_data_accuracy)
```
